```python
import numpy as np
import pandas as pd
import random 
import math
import matplotlib.pyplot as plt
from scipy.stats import norm
from sys import maxsize
```

```python
a=np.loadtxt(open("german.data-numeric", "rb"))

for i in range(0,1000):
    if a[i,24]==2:
        a[i,24]=-1


c1=list(range(0,1000))
rds=random.sample(c1,700)
rds_sort=np.sort(rds)
x=a[rds_sort,0:24]
y=a[rds_sort,24]
np.ones(700).shape
xnew=np.hstack((x,np.ones((700,1))))
```

# Linear Least Squares
    

```python
A = xnew
w_star= np.matmul(np.linalg.inv(np.matmul(A.T,A)),np.matmul(A.T,y))
```

```python
w_star
```

```python
cost=np.dot((y-np.matmul(A,w_star.T)),(y-np.matmul(A,w_star.T)))/700
cost
```

```python
yr=[]
for i in range(0,700):
    if np.dot(w_star,xnew[i,:])>0:
        yr.append(1)
    else:
        yr.append(-1)
```

```python
costr=np.dot((y-yr),(y-yr))/700
costr
```

# Logistic Regression

```python
epsi=0.0001
eta=0.0001
wk=np.zeros(25)
i=0
wkn=0.1*np.ones(25)

print("Norm",np.dot((wkn-wk),(wkn-wk)))

while np.dot((wkn-wk),(wkn-wk)) > epsi:
    
    sum1=0
    for i in range(0,700):
        ea=np.dot(wk,xnew[i,:])
        h=1/(1+math.exp(-ea))
        sum1=sum1+h*h*math.exp(-ea)*(xnew[i,:].T*(h - y[i]))
    wkn = wk
    
    wk=wk-eta*sum1
    print("Norm",np.dot((wkn-wk),(wkn-wk)))
```

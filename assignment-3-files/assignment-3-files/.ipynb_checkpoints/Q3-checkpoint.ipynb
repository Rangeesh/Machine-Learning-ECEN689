```python
import numpy as np
import pandas as pd
import random 
import math
import matplotlib.pyplot as plt
from scipy.stats import norm
from sys import maxsize
```

```python
a=np.loadtxt(open("german.data-numeric", "rb"))

for i in range(0,1000):
    if a[i,24]==2:
        a[i,24]=-1

a
c1=list(range(0,1000))
rds=random.sample(c1,700)
rds_sort=np.sort(rds)
x=a[rds_sort,0:24]
y=a[rds_sort,24]

```

```python
x.shape
```

```python
epsi=1
eta=0.01
wk=np.zeros(24)
wk1=np.ones(24)

while np.linalg.norm(wk1-wk)>epsi:
    wk=wk1
    grad_f= np.matmul(x.T,(y-np.dot(x,wk)))
    wk1=wk+eta*grad_f


    
    
    
    
```

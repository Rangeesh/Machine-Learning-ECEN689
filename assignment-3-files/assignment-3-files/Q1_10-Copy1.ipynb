```python
import numpy as np
import pandas as pd
import random 
import math
import csv
import matplotlib.pyplot as plt
from scipy.stats import norm
from sys import maxsize
```

```python
with open('Gamma_test.csv', newline='') as csvfile:
    data = list(csv.reader(csvfile))
at=(np.float_(data))
with open('Gamma_train.csv', newline='') as csvfile:
    data = list(csv.reader(csvfile))
a=(np.float_(data))

xt=at[:,0:2]
yt=at[:,2]
At=np.hstack((xt,np.ones((1000,1))))

N=500
c1=list(range(0,2000))
rds=random.sample(c1,N)

rs=np.sort(rds)
xnew=a[rs,0:2]
A=np.hstack((xnew,np.ones((N,1))))
y=a[rs,2]
```

```python
epsi=0.0001
eta=0.0001
wk=np.zeros(3)
i=0
wkn=0.1*np.ones(3)

print("Norm",np.dot((wkn-wk),(wkn-wk)))

while np.dot((wkn-wk),(wkn-wk)) > epsi:
    
    sum1=0
    for i in range(0,N):
        ea=np.dot(wk,A[i,:])
        h=1/(1+math.exp(-ea))
        sum1=sum1+h*h*math.exp(-ea)*(A[i,:].T*(h - y[i]))
    wkn = wk
    
    wk=wk-eta*sum1
    print("Norm",np.dot((wkn-wk),(wkn-wk)))
```

```python
yr=[]
for i in range(0,1000):
    if np.dot(wk,At[i,:])>0:
        yr.append(1)
    else:
        yr.append(-1)
```
